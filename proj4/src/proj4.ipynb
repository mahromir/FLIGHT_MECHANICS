```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_table("../input/char_samolot.txt", sep = "\t")
df.columns
```

```python
import numpy as np

fig, ax = plt.subplots(figsize = (9, 6))
fig.autofmt_xdate(rotation=90)

ax.set_title("Biegunowa")

ax.plot(df['V_1'], df["w_1"]*(-1), label = "$m_{max}$", linestyle = "solid", color = 'black')
ax.plot(df['V_2'], df["w_2"]*(-1), label = "$m_{min}$", linestyle = "dotted", color = 'black')
ax.plot(df['V_3'], df["w_3"]*(-1), label = "$m_{avg}$", linestyle = "dashed", color = 'black')


v_1 = np.linspace(0, 90, 10)
gamma = min(df["gamma"])
ax.plot(v_1, v_1*(-1)*np.deg2rad(gamma), label = f"$\\gamma = {round(gamma,2)}^o $", alpha = 0.9, linewidth = 0.8, color = "red")

# ax.axvline(54, label = "$V_{max}$")
# ax.axvline(64, label = "$1.2\\cdot V_{max}$")

ax.set_ylabel("$w [\\frac{m}{s}]$      ", rotation = 0)
ax.set_xlabel("$V [\\frac{m}{s}]$      ")

ax.set_xticks(np.arange(0, 61, 1))
# ax.set_xticks()


ax.set_yticks(np.arange(0, -10.5, -0.5))
ax.set_yticklabels(np.arange(0, 10.5, 0.5))


ax.grid()

# ax.set_ylim([0, max(df['w_2'])])
ax.set_xlim([0, 50])
ax.set_ylim([-8, 0])


ax.legend()
plt.tight_layout()

plt.savefig("../img/Biegunowa.jpg", dpi= 300)
plt.show()

```
