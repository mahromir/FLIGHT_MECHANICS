```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_table("../input/coef.txt")
# df
print(df.dtypes)
# print(df.columns)
df = df[5:35]
```

```python
fig,ax = plt.subplots()
ax.plot(df["cz"], df['Cmbu1'], label = "$C_{mbu1}$", linestyle = "solid", c = 'black')
ax.plot(df["cz"], df['Cmbu2'], label = "$C_{mbu2}$", linestyle = "dotted" , c = 'black')
ax.plot(df["cz"], df['Cmbu3'], label = "$C_{mbu3}$", linestyle = "dashed", c = 'black')
ax.grid(True)
ax.set_xlabel("$C_z$")
ax.legend()
plt.savefig("../output/Cmbu.jpg", dpi = 300)
```
